```python
import cv2
import numpy as np
from openvino.inference_engine import IECore

ie = IECore()
net = ie.read_network(model='C:/Users/abir.DESKTOP-8COR111/Desktop/Projet-SE/emotions-recognition-retail-0003.xml', weights='emotions-recognition-retail-0003.bin')
input_blob = next(iter(net.input_info))
n, c, h, w = net.input_info[input_blob].input_data.shape

# Load the network to the plugin
exec_net = ie.load_network(network=net, device_name="CPU")

# Open the webcam
cap = cv2.VideoCapture(0)

while True:
    # Read the current frame from the webcam
    ret, frame = cap.read()
    
    # Preprocess the input image
    resized_img = cv2.resize(frame, (w, h))
    input_data = resized_img.transpose((2, 0, 1))  # Change the layout from HWC to CHW
    input_data = input_data.reshape((n, c, h, w))  # Add a batch dimension
    
    # Run inference
    output_blob = next(iter(net.outputs))
    results = exec_net.infer(inputs={input_blob: input_data})
    emotions = results[output_blob][0]
    
    # Print the predicted emotion
    emotions_dict = {0: 'neutral', 1: 'happy', 2: 'sad', 3: 'surprise', 4: 'anger'}
    emotion_label = emotions_dict[np.argmax(emotions)]
    print(f"Predicted emotion: {emotion_label}")
    
    # Display the frame with the predicted emotion label
    cv2.putText(frame, emotion_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.imshow('Emotion Detection', frame)
    
    # Exit if the 'q' key is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```
